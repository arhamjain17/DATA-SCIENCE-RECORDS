```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.random.randint(1,20,100)
y1=x+5
y2=x**2
y3=x**3
```

```python
plt.figure(figsize=(10,3))
plt.scatter(x,y1,c="g")
plt.scatter(x,y2,c="r")
plt.scatter(x,y3,c="b")
plt.grid(True)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.scatter(x,y1,marker="*",c="g",s=100)

plt.subplot(1,2,2)
plt.scatter(x,y2,marker=".",c="r",s=200)
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(x,y1,color="g",linestyle=":",linewidth=2)
plt.plot(x,y2,color="r",linestyle="-.",linewidth=3)
plt.plot(x,y3,color="b",linestyle=".",linewidth=3)
plt.grid(True)
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.plot(x,y1,color="g",linestyle=":",linewidth=3)

plt.subplot(1,3,2)
plt.plot(x,y2,color="r",linestyle="-.",linewidth=3)

plt.subplot(1,3,3)
plt.plot(x,y3,color="b",linestyle=":",linewidth=3)

plt.show()
```

```python
my_data=np.array([-500,300,400,300,900,1900,600,1000,200])
plt.figure(figsize=(10,3))
sns.boxplot(x=my_data)
```

```python
q1=np.quantile(my_data,.25)
q3=np.quantile(my_data,.75)
iqr=q3-q1
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
lower_limit,upper_limit
```

```python
my_data[my_data>upper_limit]
```

```python
my_data[my_data<lower_limit]
```
